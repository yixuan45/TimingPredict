```python
import numpy as np
import matplotlib.pyplot as plt

# 设置随机种子，确保结果可重现
np.random.seed(42)
plt.rcParams['font.sans-serif'] = ['SimHei']
plt.rcParams['axes.unicode_minus'] = False

# 生成随机游走数据
def generate_random_walk(steps=100):
    """生成随机游走时间序列"""
    # 生成随机增量（正态分布，均值为0，标准差为1）
    increments = np.random.normal(0, 1, steps)

    # 计算累积和（随机游走）
    random_walk = np.cumsum(increments)

    # 添加初始值0
    random_walk = np.insert(random_walk, 0, 0)

    # 生成时间点
    time_points = np.arange(0, steps + 1)

    return time_points, random_walk

# 绘制单条随机游走序列
def plot_single_random_walk(steps=100, title="随机游走时间序列"):
    """绘制单条随机游走序列"""
    plt.figure(figsize=(10, 6))

    # 生成随机游走数据
    time, walk = generate_random_walk(steps)

    # 绘制随机游走
    plt.plot(time, walk, 'b-', linewidth=2, alpha=0.7)

    # 添加零水平线作为参考
    plt.axhline(y=0, color='r', linestyle='--', alpha=0.3)

    # 标记起点和终点
    plt.scatter(time[0], walk[0], color='green', s=100, label='起点')
    plt.scatter(time[-1], walk[-1], color='red', s=100, label='终点')

    plt.title(title, fontsize=14)
    plt.xlabel('时间步', fontsize=12)
    plt.ylabel('随机游走值', fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.legend()
    plt.tight_layout()
    plt.show()

# 运行函数绘制随机游走
plot_single_random_walk(steps=200)
```
