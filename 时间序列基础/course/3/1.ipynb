```python
import numpy as np
import matplotlib.pyplot as plt

# 生成白噪声序列，长度为 200
np.random.seed(0)  # 设置随机种子，保证结果可复现
e = np.random.normal(0, 1, 200)

# 初始化四个时间序列
y1 = np.zeros(200)
y2 = np.zeros(200)
y3 = np.zeros(200)
y4 = np.zeros(200)

# 给第一个元素赋值
y1[0] = e[0]
y2[0] = e[0]
y3[0] = e[0]
y4[0] = e[0]

# 模拟 AR(1) 过程
for i in range(1, 200):
    y1[i] = 0.5 * y1[i - 1] + e[i]
    y2[i] = 0.5 * y2[i - 1] + e[i]
    y3[i] = 0.5 * y3[i - 1] + e[i]
    y4[i] = 0.5 + 0.2 * i + 0.5 * y4[i - 1] + e[i]

# 绘制 y1 的折线图
plt.figure(figsize=(5, 6))
plt.plot(y1)  # 移除了错误的 type='l' 参数
plt.title('AR(1) Process for y1')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```

```python
plt.figure(figsize=(5, 6))
plt.plot(y2)  # 移除了错误的 type='l' 参数
plt.title('AR(1) Process for y2')
plt.xlabel('Time')
plt.ylabel('Value')
plt.show()
```
